```python
import yfinance as yf
import pandas as pd
```

```python
august = yf.download('SPY', start='2024-08-01', end='2024-08-31', interval = '5m')
print(august.head(30))
print(august.columns)
```

```python
def dt(val):
    value_str = str(val)
    first25 = value_str[:25]
    return first25
```

```python
august['datetime'] = august['Open'].apply(dt)
august = august.drop(columns=['datetime'])
```

```python
import matplotlib as mpl
import mplfinance as mpf

mpf.plot(august, type='candle', style='charles', volume=True)
```

```python
morning = august.between_time('09:30:00', '11:00:00')
morning.head(50)
```

```python


mpf.plot(morning, type='candle', style='charles', volume=True)
```

```python
def reversion_changes(df):
    filtered = df.between_time('10:00:00', '10:30:00')
    results = []
    for date, group in filtered.groupby(filtered.index.date):
        open_price = group.iloc[0]['Open']
        close_price = group.iloc[-1]['Close']
        dollar_change = close_price - open_price
        percent_change = (dollar_change / open_price) * 100
        results.append({'Date': date, 'Dollar Change': dollar_change, 'Percent Change': percent_change})
    return pd.DataFrame(results)

aug_sep_td = yf.download('SPY', start='2024-08-01', end='2024-09-13', interval='5m')


def morning_changes(df):
    filtered = df.between_time('09:45:00', '10:00:00')
    results = []
    for date, group in filtered.groupby(filtered.index.date):
        open_price = group.iloc[0]['Open']
        close_price = group.iloc[-1]['Close']
        dollar_change = close_price - open_price
        percent_change = (dollar_change / open_price) * 100
        results.append({'Date': date, 'Dollar Change': dollar_change, 'Percent Change': percent_change})
    return pd.DataFrame(results)
print(reversion_changes(aug_sep_td))

```

```python

morning_df = morning_changes(aug_sep_td)
reversion_df = reversion_changes(aug_sep_td)

# Calculate the correlation between the percent changes of morning and reversion
correlation = morning_df['Percent Change'].corr(reversion_df['Percent Change'])
print(f"Correlation between morning and reversion percent changes: {correlation}")

```

```python
# Filter the data for the date range September 3-6, 2024
import matplotlib.dates as mdates

sep_3_6 = aug_sep_td.loc['2024-09-03':'2024-09-06']

# Plot the filtered data using mplfinance
fig, axlist = mpf.plot(sep_3_6, type='candle', style='charles', volume=True, returnfig=True)

# Highlight the time ranges 9:45-10:00 and 10:00-10:30 each day
for date in pd.date_range(start='2024-09-03', end='2024-09-06', freq='B'):
    start_945 = pd.Timestamp(f'{date.date()} 09:45:00')
    end_1000 = pd.Timestamp(f'{date.date()} 10:00:00')
    start_1000 = pd.Timestamp(f'{date.date()} 10:00:00')
    end_1030 = pd.Timestamp(f'{date.date()} 10:30:00')
    
    axlist[0].axvspan(start_945, end_1000, color='yellow', alpha=0.3)
    axlist[0].axvspan(start_1000, end_1030, color='red', alpha=0.3)

# Format the x-axis to show the dates and times clearly
axlist[0].xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d %H:%M'))
plt.setp(axlist[0].xaxis.get_majorticklabels(), rotation=45)
plt.show()
```

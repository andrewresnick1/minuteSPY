```python
import yfinance as yf
import pandas as pd
```

```python
august = yf.download('SPY', start='2024-08-01', end='2024-08-31', interval = '5m')
print(august.head(30))
print(august.columns)
```

```python
def dt(val):
    value_str = str(val)
    first25 = value_str[:25]
    return first25
```

```python
august['datetime'] = august['Open'].apply(dt)
august = august.drop(columns=['datetime'])
```

```python
import matplotlib as mpl
import mplfinance as mpf

mpf.plot(august, type='candle', style='charles', volume=True)
```

```python
def reversion_changes(df):
    filtered = df.between_time('10:00:00', '10:30:00')
    results = []
    for date, group in filtered.groupby(filtered.index.date):
        open_price = group.iloc[0]['Open']
        close_price = group.iloc[-1]['Close']
        dollar_change = close_price - open_price
        percent_change = (dollar_change / open_price) * 100
        results.append({'Date': date, 'Dollar Change': dollar_change, 'Percent Change': percent_change})
    return pd.DataFrame(results)

aug_sep_td = yf.download('SPY', start='2024-08-01', end='2024-09-13', interval='5m')


def morning_changes(df):
    filtered = df.between_time('09:45:00', '10:00:00')
    results = []
    for date, group in filtered.groupby(filtered.index.date):
        open_price = group.iloc[0]['Open']
        close_price = group.iloc[-1]['Close']
        dollar_change = close_price - open_price
        percent_change = (dollar_change / open_price) * 100
        results.append({'Date': date, 'Dollar Change': dollar_change, 'Percent Change': percent_change})
    return pd.DataFrame(results)
print(reversion_changes(aug_sep_td))

```

```python

morning_df = morning_changes(aug_sep_td)
reversion_df = reversion_changes(aug_sep_td)

# Calculate the correlation between the percent changes of morning and reversion
correlation = morning_df['Percent Change'].corr(reversion_df['Percent Change'])
print(f"Correlation between morning and reversion percent changes: {correlation}")

```

```python
def price_movement(df):
    results = []
    for date, group in df.groupby(df.index.date):
        data_955_1000 = group.between_time('09:55:00', '10:00:00')
        data_1000_1030 = group.between_time('10:00:00', '10:30:00')
        
        if not data_955_1000.empty and not data_1000_1030.empty:
            open_955 = data_955_1000.iloc[0]['Open']
            close_1000 = data_955_1000.iloc[-1]['Close']
            change_955_1000 = close_1000 - open_955
            direction_955_1000 = 'Upward' if change_955_1000 > 0 else 'Downward'
            
            open_1000 = data_1000_1030.iloc[0]['Open']
            close_1030 = data_1000_1030.iloc[-1]['Close']
            change_1000_1030 = close_1030 - open_1000
            direction_1000_1030 = 'Upward' if change_1000_1030 > 0 else 'Downward'
            
            results.append({
                'Date': date,
                '9:55-10:00 Direction': direction_955_1000,
                '9:55-10:00 Change': change_955_1000,
                '10:00-10:30 Direction': direction_1000_1030,
                '10:00-10:30 Change': change_1000_1030
            })
    
    return pd.DataFrame(results)

price_movement_df = price_movement(aug_sep_td)
print(price_movement_df)

# Calculate the frequency and magnitude of direction deviations
deviation_count = 0
total_deviation_change = 0
non_deviation_count = 0
total_non_deviation_change = 0

for index, row in price_movement_df.iterrows():
    if row['9:55-10:00 Direction'] != row['10:00-10:30 Direction']:
        deviation_count += 1
        total_deviation_change += abs(row['9:55-10:00 Change'] - row['10:00-10:30 Change'])
    else:
        non_deviation_count += 1
        total_non_deviation_change += abs(row['9:55-10:00 Change'] - row['10:00-10:30 Change'])

if deviation_count > 0:
    average_deviation_change = total_deviation_change / deviation_count
else:
    average_deviation_change = 0

if non_deviation_count > 0:
    average_non_deviation_change = total_non_deviation_change / non_deviation_count
else:
    average_non_deviation_change = 0

print(f"Number of deviations: {deviation_count}")
print(f"Average price change during deviations: {average_deviation_change}")
print(f"Number of non-deviations: {non_deviation_count}")
print(f"Average price change during non-deviations: {average_non_deviation_change}")

import matplotlib.pyplot as plt
# Create a distribution table for direction changes
distribution_table = price_movement_df.groupby(['Date', '9:55-10:00 Direction', '10:00-10:30 Direction']).size().unstack(fill_value=0)
print("Distribution Table:")
print(distribution_table)

# Plot the distribution table
fig, ax = plt.subplots(figsize=(12, 8))
distribution_table.plot(kind='bar', stacked=True, ax=ax)
ax.set_title('Distribution of Direction Changes')
ax.set_xlabel('Date')
ax.set_ylabel('Count')
ax.legend(title='Direction from 10:00 to 10:30')
plt.xticks(rotation=45)
plt.show()

# Calculate the average change for each direction combination
average_change_table = price_movement_df.groupby(['Date', '9:55-10:00 Direction', '10:00-10:30 Direction'])[['9:55-10:00 Change', '10:00-10:30 Change']].mean()
print("\nAverage Change Table:")
print(average_change_table)

# Plot the average change table
fig, ax = plt.subplots(figsize=(12, 8))
average_change_table.plot(kind='bar', ax=ax)
ax.set_title('Average Change for Each Direction Combination')
ax.set_xlabel('Date')
ax.set_ylabel('Average Change')
ax.legend(title='Time Period')
plt.xticks(rotation=45)
plt.show()
```
